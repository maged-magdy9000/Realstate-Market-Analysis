```python
df['price of square m']= round(df['price']/df['area'],1)
df.head()
```

```python
df['state']=df['location'].apply(lambda x:x.split(',')[0])
df['country']=df['location'].apply(lambda x:x.split(',')[1:2])
df['District']=df['location'].apply(lambda x:x.split(',')[-1])
```

```python
df['Unfurnished']=df['amenities'].apply(lambda x:1 if 'unfurnished' in x.lower() else 0)
df['Balcony']=df['amenities'].apply(lambda x:1 if 'balcony' in x.lower() else 0)
df['Private Garden']=df['amenities'].apply(lambda x:1 if 'private garden' in x.lower() else 0)
df['Covered Parking']=df['amenities'].apply(lambda x:1 if 'covered parking' in x.lower() else 0)
df['Security']=df['amenities'].apply(lambda x:1 if 'security' in x.lower() else 0)
```
